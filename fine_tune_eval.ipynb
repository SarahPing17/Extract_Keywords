```python
def get_result_all_major_content(file_path):
    # Open and read the JSON file
    with open(file_path, 'r') as file:
        data = json.load(file)

    # Extract the 'jsonFile' and 'keyword' values
    json_file_content = data.get('jsonFile', '')
    keywords = data.get('keyword', '')

    completion = client.chat.completions.create(
        model="ft:gpt-3.5-turbo-0125:personal::9C9ARGC3",
        messages=[
            {"role": "system", "content": "You are a helpful assistant who will help me to summarize the keywords from the given research papers."},
            {"role": "user", "content": json_file_content}
            ]
            )
    model_return = completion.choices[0].message.content
    print(model_return)

    # Update the 'jsonFile' content in the data
    data['model_result'] = model_return

    # Write the modified JSON data back to the file
    with open(file_path, 'w') as file:
        json.dump(data, file, indent=4)
```
